```python
import cifar10
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
cifar10.data_path = "data/CIFAR-10/"
```

```python
cifar10.maybe_download_and_extract()
```

```python
class_names = cifar10.load_class_names()
class_names
```

```python
images_train, cls_train, labels_train = cifar10.load_training_data()
images_test, cls_test, labels_test = cifar10.load_test_data()
```

```python
pca=PCA()
```

```python
print(images_train.shape)
```

```python
print(labels_train)
```

```python
images_train=images_train.reshape([50000,3072])
```

```python
images_train.shape
```

```python
k=0
total=0
while total<0.99:
    total=total+pca.explained_variance_ratio_[k]
    k=k+1
k
```

```python
#658 is the number of features pca has reduced the components to.
```

```python
pca_updated=PCA(n_components=658,whiten=True)
```

```python
images_transformed=pca_updated.fit_transform(images_train)
images_transformed.shape
```

```python
images_test.shape
```

```python
images_test=images_test.reshape([10000,3072])
```

```python
images_test_transformed=pca_updated.transform(images_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
cls_train.shape
```

```python
rf.fit(images_transformed,cls_train)
```

```python
cls_predict=rf.predict(images_test_transformed)
```

```python
cls_predict
```

```python
from sklearn.metrics import classification_report
print(classification_report(cls_predict,cls_test))
```

```python
#mapping numbers of classes back to class names and storing in csv
class_names
```

```python
classNameDict={}
i=0
for c in class_names:
    classNameDict[i]=class_names[i]
    i+=1
classNameDict
```

```python
cls_predict_names=[]
for c in cls_predict:
    cls_predict_names.append(classNameDict[c])
    c+=1
cls_predict_names
```
